```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df_numeric = df.select_dtypes(include=[np.number])
df_numeric = df_numeric.fillna(df_numeric.median())

kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(df_numeric)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_numeric)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=clusters, cmap='viridis', s=50, alpha=0.5)
plt.title('Clusterização das casas')
plt.xlabel('Principal 1')
plt.ylabel('Principal 2')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=42)
df_tsne = tsne.fit_transform(df_numeric)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_tsne[:, 0], df_tsne[:, 1], c=clusters, cmap='viridis', s=50, alpha=0.5)
plt.title('Visualização das características das casas')
plt.xlabel('Componente 1 do t-SNE')
plt.ylabel('Componente 2 do t-SNE')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```
